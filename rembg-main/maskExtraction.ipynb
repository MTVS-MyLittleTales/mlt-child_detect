```python
from rembg import remove
import numpy as np
import cv2

def bg_remove(input_path, output_path, write=0):
    input_path = input_path
    output_path = output_path

    input = cv2.imread(input_path)
    output = remove(input)
    if write:
        cv2.imwrite(output_path, output)
    return output

def create_mask(output, output_path, write=0):
    mask = output[:, :, -1]
    mask = cv2.cvtColor(mask, cv2.COLOR_GRAY2RGBA)
    if write:
        cv2.imwrite(output_path, mask)

    return mask

input_path = "C:/Users/HP/Desktop/backgroundimg/input/child2.jpg"
output_path = 'testoutput/output.png'
output = bg_remove(input_path=input_path, output_path=output_path)
mask_img = create_mask(output=output, output_path=output_path, write=1)
```
